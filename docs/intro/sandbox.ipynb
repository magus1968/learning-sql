## div block

division block – секция (разделение, отделение);  
див-блок (или блок-контейнер) – служит "коробкой" для группировки текста, картинок или других блоков.

Шкала размеров текста в блоках `:::div` формируется через `:class:`

```markdown
:::{div}
:class: lead
`lead` используется для лид-абзацев(1.25rem)
:::
```
